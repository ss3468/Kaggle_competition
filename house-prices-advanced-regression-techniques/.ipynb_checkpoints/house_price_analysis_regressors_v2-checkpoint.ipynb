```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data=train_data[['Neighborhood','LotArea','LotShape','LotConfig','LandContour','HouseStyle','LotFrontage']]
training, validation = train_test_split(data, test_size=0.2, random_state=42)
```

```python
from scipy.stats import skew
def choose_agg_method(series):
    if abs(skew(series.dropna())) > 1:  # Consider it skewed if |skewness| > 1
        return series.median()
    else:
        return series.mean()
train_data_imputed_simple = training.copy()
def handle_outliers(df):
    # Calculate the IQR and bounds within each group
    Q1 = df['LotFrontage'].quantile(0.25)
    Q3 = df['LotFrontage'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Clip the values outside the bounds
    df['LotFrontage'] = df['LotFrontage'].clip(lower_bound, upper_bound)
    return df

# Apply the function to each group
train_data_imputed_simple = train_data_imputed_simple.groupby('Neighborhood').apply(handle_outliers).reset_index(drop=True)
grouped_stats = train_data_imputed_simple.groupby('Neighborhood')['LotFrontage'].agg(
    mean='mean',
    median='median',
    skewness=lambda x: skew(x.dropna())
).reset_index()
def choose_imputation_value(row):
    if abs(row['skewness']) > 1:  # Highly skewed
        return row['median']
    else:
        return row['mean']

# Add a column for chosen imputation value
grouped_stats['imputation_value'] = grouped_stats.apply(choose_imputation_value, axis=1)
grouped_stats
```

```python
imputation_map = grouped_stats.set_index('Neighborhood')['imputation_value']
train_data_imputed_simple['LotFrontage'] = train_data_imputed_simple.apply(
    lambda row: imputation_map[row['Neighborhood']] if pd.isnull(row['LotFrontage']) else row['LotFrontage'], 
    axis=1
)
training
```

```python
from sklearn.impute import SimpleImputer
# simple_imputer = SimpleImputer(strategy='mean')
# training['LotFrontage'] = simple_imputer.fit_transform(training[['LotFrontage']])
X_train_imputed = train_data_imputed_simple.drop(columns=['LotFrontage'])
X_train_imputed['LotArea']=np.sqrt(X_train_imputed['LotArea'])
y_train_imputed = np.log1p(train_data_imputed_simple['LotFrontage'])
X_valid = validation.drop(columns=['LotFrontage'])
X_valid['LotArea']=np.sqrt(X_valid['LotArea'])
y_valid =np.log1p(validation['LotFrontage'])
```

```python
from sklearn.preprocessing import OneHotEncoder
columns_to_encode = ['LotShape','LotConfig','LandContour','HouseStyle']
encoder = OneHotEncoder(drop='first',sparse=False, handle_unknown='ignore')
X_train_encoded = encoder.fit_transform(X_train_imputed[columns_to_encode])
X_valid_encoded = encoder.transform(X_valid[columns_to_encode])
X_train_encoded_df = pd.DataFrame(X_train_encoded, columns=encoder.get_feature_names_out(columns_to_encode))
X_valid_encoded_df = pd.DataFrame(X_valid_encoded, columns=encoder.get_feature_names_out(columns_to_encode))
X_train = X_train_imputed.drop(columns=columns_to_encode).reset_index(drop=True).join(X_train_encoded_df)
X_valid = X_valid.drop(columns=columns_to_encode).reset_index(drop=True).join(X_valid_encoded_df)
frequency_encoding = X_train['Neighborhood'].value_counts() / len(X_train)
# Neighborhood_mapper={'Blmngtn':1, 'Blueste':2, 'BrDale':3, 'BrkSide':4, 'ClearCr':5, 'CollgCr':6, 'Crawfor':7, 'Edwards':8, 'Gilbert':9, 'IDOTRR':10, 'MeadowV':11, 'Mitchel':12, 'NAmes':13, 'NoRidge':14, 'NPkVill':15, 'NridgHt':16, 'NWAmes':17, 'OldTown':18, 'SWISU':19, 'Sawyer':20, 'SawyerW':21, 'Somerst':22, 'StoneBr':23, 'Timber':24, 'Veenker':25}
X_train['Neighborhood']=X_train['Neighborhood'].map(frequency_encoding)#.map(Neighborhood_mapper)
X_valid['Neighborhood']=X_valid['Neighborhood'].map(frequency_encoding)#.map(Neighborhood_mapper)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(random_state=42)#LinearRegression()
model.fit(X_train,y_train_imputed)
y_pred=model.predict(X_valid)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
non_null_mask = ~y_valid.isnull()
y_test_non_null = np.expm1(y_valid[non_null_mask])
y_pred_non_null = np.expm1(y_pred[non_null_mask])
# Evaluate using non-null y_test
mse = mean_squared_error(y_test_non_null, y_pred_non_null)
mae = mean_absolute_error(y_test_non_null, y_pred_non_null)
r2 = r2_score(y_test_non_null, y_pred_non_null)

print(f"MSE: {mse}, MAE: {mae}, R²: {r2}")
```

```python
# Get feature importance
importances = model.feature_importances_

# Sort feature importance in descending order
indices = importances.argsort()

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.title('Feature Importance')
plt.barh(range(X_train.shape[1]), importances[indices], align='center')
plt.yticks(range(X_train.shape[1]), [X_train.columns[i] for i in indices])
plt.xlabel('Feature Importance')
plt.show()
```

```python
from bs4 import BeautifulSoup
import re
import unicodedata
def convert_enconcoded_html(text):
    tem=BeautifulSoup(text, 'html.parser').get_text()
    return tem
def remove_html_tags_func(text):
    tem=re.sub(r"<[^>]+>",' ',text)
    tem=re.sub(r"<[^>]+\.{1,} ",' ',tem)
    tem=re.sub(r"\&amp;",'&',tem)
    return tem
def remove_url_func(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)
def remove_tweet_funct(text):
    text = re.sub(r'RT(\s?)\@\w+','',text)
    text = re.sub(r'RT\:\s\@\w+','',text)
    text = re.sub(r'RT\s\#\w+','',text)
    text = re.sub(r'RT\:\s\#\w+','',text)
    text = re.sub(r'RT\:\s\w+','',text)
    text = re.sub(r'[tT]hx\s\@\w+','',text)
    text = re.sub(r'MT\s\@\w+','',text)
    text = re.sub(r'DT\s\@\w+','',text)
    text = re.sub(r'\.\@\w+','',text)
    text = re.sub(r'\(via\s\@\w+\)','',text)
    text = re.sub(r'\B((\@\S+)|#)', '', text) #removes @mentions #removes hastag '#' symbol
    #text = re.sub('RT[\s]+','',text)
    text = re.sub('\n',' ',text)
    return text
def remove_accented_chars_func(text):
    return ''.join([c for c in text if ord(c) < 128])#unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
def clean_repeating_char(text):
    text=re.sub(r'([a-zA-Z])\1\1+', r'\1\1', text,flags=re.I)#(.)\1{3,}
    text=re.sub(r'(\B:)(3)+', r'\1\2', text,flags=re.I)
    text = re.sub(r'\blo+l\b',"lol",text,flags=re.I)
    text = re.sub(r'go+a+l+','goal',text)
    text = re.sub(r'\bso+\b',"so",text,flags=re.I)
    text = re.sub(r'\bno+\b',"no",text,flags=re.I)
    text = re.sub(r'\bo+h+o*\b','oh',text,flags=re.I)
    text = re.sub(r'\ba+\b','a',text,flags=re.I)
    text = re.sub(r'\ba+ll\b','all',text,flags=re.I)
    text = re.sub(r'\ba+n+\b','an',text,flags=re.I)
    text = re.sub(r'\ba+n+d+\b','and',text,flags=re.I)
    text=re.sub(r'\bab+andon','abandon',text,flags=re.I)
    text=re.sub(r'\blmfa(o){2,}(\?|\.?)','lmfao',text,flags=re.I)
    return text
emoticons=[r"\B\(\:\B",r"\B(\>?)\:3\b",r"<3\b",r"(\B>_<\B)|(\B>_>\B)",r"\B-(_+)-\B",r"\B:(-?)D\b",r"\B:(-?)o\b",r"\B:(-?)\){1,}\B",r"\B:(-?)\({1,}\B",r"\B;(-?)\)\B",r"\bxD\b",r"\B:[oO]\b",r"\B\;\-\;\B",r"\B\:[Pp]\b",r"\B=/\B",r"\B\:[Ss]\b",r'\:/\B',r"\B\;3\b"]
def remove_emoticon(text):
    m=r""+"|".join(emoticons)
    text = re.sub(m," ", text)
    return text
def clean_text(text):
    text=convert_enconcoded_html(text)
    text=remove_html_tags_func(text)
    text=remove_url_func(text)
    text=remove_tweet_funct(text)
    text=remove_accented_chars_func(text)
    text=clean_repeating_char(text)
    text=remove_emoticon(text)
    return text
stop_words = set(stopwords.words('english'))
wn = nltk.WordNetLemmatizer()
worddict = set(nltk.corpus.words.words())
def norm_lemm_v_a_func(text):
    words1 = word_tokenize(text)
    text1 = ' '.join([wn.lemmatize(word, pos='v') for word in words1])
    words2 = word_tokenize(text1)
    text2 = ' '.join([wn.lemmatize(word, pos='a') for word in words2])
    return text2
def preprocessing(text):
    new_text = clean_text(text)
    tokens = word_tokenize(new_text.lower().strip())
    new_text=" ".join([token for token in tokens if token not in stopwords.words("english")])
    return norm_lemm_v_a_func(new_text)
```

```python
data_train['clean_text']=data_train['text'].apply(lambda x: clean_text(x))
data_test['clean_text']=data_test['text'].apply(lambda x: clean_text(x))
data_train['new_text']=data_train['clean_text'].apply(lambda x: preprocessing(x))
data_test['new_text']=data_test['clean_text'].apply(lambda x: preprocessing(x))
data_train.to_csv('tem_train.csv',encoding='ISO-8859–1',index=False)
data_test.to_csv('tem_test.csv',encoding='ISO-8859–1',index=False)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    data_train['new_text'],
    data_train['target'],stratify=data_train['target'],
    test_size=0.25,
    random_state=2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics
tf_transformer=TfidfVectorizer(use_idf=True,sublinear_tf=True)
x_train_tf = tf_transformer.fit_transform(x_train)
y_train = np.array(y_train)
x_pre_tf = tf_transformer.transform(x_test)
x_test_tf = tf_transformer.transform(data_test['new_text'])
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
from sklearn.metrics import confusion_matrix
MNB_model = MultinomialNB(alpha=1.05, fit_prior=True, class_prior=None)
MNB_model.fit(x_train_tf, y_train)
predicted = MNB_model.predict(x_pre_tf)
c5=metrics.classification_report(list(y_test), predicted)
print(c5)
```

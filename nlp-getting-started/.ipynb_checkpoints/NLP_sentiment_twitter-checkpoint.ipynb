```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
from nltk.corpus import stopwords
from nltk.corpus import words
from nltk.tokenize import word_tokenize
from nltk.corpus.reader.wordnet import *
import re
import contractions
import string
from sklearn.metrics import classification_report
```

```python
from bs4 import BeautifulSoup
def convert_enconcoded_html(text):
    tem=BeautifulSoup(text, 'html.parser').get_text()
    return tem
def remove_html_tags_func(text):
    tem=re.sub(r"<[^>]+>",' ',text)
    tem=re.sub(r"<[^>]+\.{1,} ",' ',tem)
    tem=re.sub(r"\&amp;",'&',tem)
    return tem
def remove_url_func(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)
def remove_retweet_label(text):
    text=re.sub(r'RT @([\w_]+)\:','', text)
    text = re.sub(r'RT(\s?)\@\w+','',text)
    text = re.sub(r'RT\:\s\@\w+','',text)
    text = re.sub(r'RT\s\#\w+','',text)
    text = re.sub(r'RT\:\s\#\w+','',text)
    text = re.sub(r'RT\:\s\w+(\:)?','',text)
    text = re.sub(r'[tT]hx\s\@\w+','',text)
    text = re.sub(r'MT\s\@\w+','',text)
    text = re.sub(r'DT\s\@\w+','',text)
    return text
def remove_twitter_handles(tweet):
    # Define a regex pattern for detecting Twitter handles and handling optional colons
    handle_pattern = re.compile(r'(?<!\w)@([A-Za-z0-9_]+)(?<!:)(?::)?(?!\d{1,2}:\d{2})')

    # Replace detected Twitter handles with an empty string
    cleaned_tweet = re.sub(handle_pattern, '', tweet)

    return cleaned_tweet
def remove_hastag(text):
    text=re.sub(r'(\w+)\#', r'\1 ', text)
    text=re.sub(r'\#', '', text)
    return text
def remove_tweet_funct(text):
    text=remove_retweet_label(text)
    text=remove_twitter_handles(text)
    text= remove_hastag(text)
    text = re.sub('\n',' ',text)
    return text
def remove_video_label(text):
    return re.sub('VIDEO:','', text)
def remove_accented_chars_func(text):
    return ''.join([c for c in text if ord(c) < 128])#unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
def clean_repeating_char(text):
    text=re.sub(r'([a-zA-Z])\1\1+', r'\1\1', text,flags=re.I)#(.)\1{3,}
    text=re.sub(r'(\B:)(3)+', r'\1\2', text,flags=re.I)
    text = re.sub(r'\blo+l\b',"lol",text,flags=re.I)
    text = re.sub(r'go+a+l+','goal',text)
    text = re.sub(r'\bso+\b',"so",text,flags=re.I)
    text = re.sub(r'\bno+\b',"no",text,flags=re.I)
    text = re.sub(r'\bo+h+o*\b','oh',text,flags=re.I)
    text = re.sub(r'\ba+\b','a',text,flags=re.I)
    text = re.sub(r'\ba+ll\b','all',text,flags=re.I)
    text = re.sub(r'\ba+n+\b','an',text,flags=re.I)
    text = re.sub(r'\ba+n+d+\b','and',text,flags=re.I)
    text=re.sub(r'\bab+andon','abandon',text,flags=re.I)
    text=re.sub(r'\blmfa(o){2,}(\?|\.?)','lmfao',text,flags=re.I)
    return text
# def remove_emoticons(text):
#     # Define a regular expression pattern to match emoticons not within numbers
#     emoticon_pattern = r'(?<!\d)[:;=]-?[\)D\(\]/\\OopP]|[\(DpPO/\\]-?[:;=]|<3|(>)?:3\b|;_;|>_<|\.:'
#     # Remove emoticons using regex
#     text_without_emoticons = re.sub(emoticon_pattern, '', text)
#     return text_without_emoticons
emoticons=[r"\:(-)?(\){1,}|\({1,}|\]|\}|/)",r"\:(-)?([Oo]|[Dd]|[Ss]|[Pp])\b",r"(>)?:3\b",r"<3\b",r"\b[xX](-)?[dD]\b",r"\>\_(\<|\>)",r";-;",r"\-(\_+)\-",r"=/(=)?",r";(-)?\)",r"[Oo]_[Oo]",r"[Tt]_[Tt]",r"\B\.\:\B"]
def remove_emoticons(text):
    m=r""+"|".join(emoticons)
    text = re.sub(m," ", text)
    return text
slang_expressions = pd.read_csv('slang_words_new.csv', sep=',')
slang_expressions_dict = dict(zip(slang_expressions.Slang_key, slang_expressions.Slang_value))
def slang_cleaner(text):
    for key in slang_expressions_dict:
        text=re.sub((r""+key),(r""+slang_expressions_dict[key]),text)
    return text
def clean_unique_slang_words(text):
    text = re.sub(r"\bJap_n\b", "Japan", text)
    text = re.sub(r"recentlu", "recently", text)
    text = re.sub(r"Ph0tos", "Photos", text)
    text = re.sub(r"amirite", "am I right", text)
    text = re.sub(r"exp0sed", "exposed", text)
    text = re.sub(r"amageddon", "armageddon", text)
    text = re.sub(r"TRAUMATISED", "traumatized", text)
    text = re.sub(r"2k([0-9])([0-9])", r"20\1\2", text,flags=re.I)
    text = re.sub(r"(\bVs\.\B)|(\b[vV][sS]\b)", r"versus", text)
    text = re.sub(r"\bS3XLEAK",'SEX LEAK',text)
    text=re.sub(r'\s*&\s*', ' and ', text)
    text=re.sub(r"\bbe4\b","before",text)
    text = re.sub(r"glowng", "glowing", text)
    text = re.sub(r"kindlng", "kindling", text)
    text = re.sub(r"riggd", "rigged", text)
    text = re.sub(r"P_EOPLE", "PEOPLE", text)
    text = re.sub(r"kostumes", "costumes", text)
    text = re.sub(r"kombatants", "combatants", text)
    text=re.sub(r"\bb4federal\b",r"before federal",text)
    text = re.sub(r"\bSuru\b", "Suruc", text)
    text = re.sub(r"\btil_now\b", "until now", text)
    text = re.sub(r"\b4PlayThursdays\b", "Foreplay Thursdays", text)
    text = re.sub(r"\bTGF2015\b", "Tontitown Grape Festival", text)
    text = re.sub(r"\bpdx911\b", "Portland Police", text)
    text = re.sub(r"\bPolitifiact\b", "PolitiFact", text)
    text = re.sub(r"\b16yr\b", "16 year", text)
    text = re.sub(r"\bKowing\b", "Knowing", text)
    text = re.sub(r"\bs2g\b", "swear to god", text)
    return text
def clean_contractions(text):
    text=contractions.fix(text)
    return text
def remove_punctuation_func(text):
    text=re.sub(r"(\d+),(\d+)",r'\1\2',text)
    return re.sub(r'[^a-zA-Z0-9]', ' ', text)
chat_expressions = pd.read_csv('chat_expressions.csv', sep=',')
chat_expressions_dict = dict(zip(chat_expressions.Chat_Words, chat_expressions.Chat_Words_Extended))
def chat_words_to_norm_words_func(text):
    return re.sub(r'\S+', lambda m: chat_expressions_dict.get(m.group().upper(), m.group()) , text)
def remove_extra_whitespaces_func(text):
    return re.sub(r'^\s*|\s\s*', ' ', text).strip()
def clean_text(text):
    text=convert_enconcoded_html(text)
    text=remove_html_tags_func(text)
    text=remove_url_func(text)
    text=remove_tweet_funct(text)
    text=remove_video_label(text)
    text=remove_accented_chars_func(text)
    text=clean_repeating_char(text)
    text=remove_emoticons(text)
    text=slang_cleaner(text)
    text=clean_unique_slang_words(text)
    text=clean_contractions(text)
#     text=remove_punctuation_func(text)
#     text=chat_words_to_norm_words_func(text)
#     text=remove_extra_whitespaces_func(text)
    return text
stop_words = set(stopwords.words('english'))
wn = nltk.WordNetLemmatizer()
worddict = set(nltk.corpus.words.words())
def norm_func(text):
    words1 = word_tokenize(text)
    text1 = ' '.join([wn.lemmatize(word) for word in words1])
    return text1
def preprocessing(text):
    new_text = clean_text(text)
    tokens = word_tokenize(new_text.lower().strip())
    new_text=" ".join([token for token in tokens if token not in stop_words])
    return norm_func(new_text)
```

```python
data_train['clean_text']=data_train['text'].apply(lambda x: clean_text(x))
data_test['clean_text']=data_test['text'].apply(lambda x: clean_text(x))
data_train['new_text']=data_train['clean_text'].apply(lambda x: preprocessing(x))
data_test['new_text']=data_test['clean_text'].apply(lambda x: preprocessing(x))
data_train.to_csv('tem_train.csv',encoding='ISO-8859–1',index=False)
data_test.to_csv('tem_test.csv',encoding='ISO-8859–1',index=False)
```

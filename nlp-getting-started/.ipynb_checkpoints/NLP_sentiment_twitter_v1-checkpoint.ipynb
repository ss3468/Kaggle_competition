```python
def remove_html(text):
    html = re.compile(r'<.*?>|&([A-Za-z0-9]+|#[0-9]{1,6}|#x[0-9a-fA-F]{1,6});')
    return re.sub(html, '', text)
def remove_url_func(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)
def remove_tweet_funct(text):
    text = re.sub('RT(\s?)\@\w+','',text)
    text = re.sub('MT\s\@\w+','',text)
    text = re.sub('\.\@\w+','',text)
    text = re.sub(r'\@\w+|\#', '', text) #removes @mentions #removes hastag '#' symbol
    #text = re.sub('RT[\s]+','',text)
    text = re.sub('\n',' ',text)
    return text
def remove_accented_chars_func(text):
    return ''.join([c for c in text if ord(c) < 128])#unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
def clean_repeating_char(text):
    text=re.sub(r'([a-zA-Z])\1\1+', r'\1\1', text,flags=re.I)#(.)\1{3,}
    text=re.sub(r'(\B:)(3)+', r'\1\2', text,flags=re.I)
    text = re.sub(r'\blo+l\b',"lol",text,flags=re.I)
    text = re.sub(r'go+a+l+','goal',text)
    text = re.sub(r'\bso+\b',"so",text,flags=re.I)
    text = re.sub(r'\bno+\b',"no",text,flags=re.I)
    text = re.sub(r'\bo+h+o*\b','oh',text,flags=re.I)
    text = re.sub(r'\ba+\b','a',text,flags=re.I)
    text = re.sub(r'\ba+ll\b','all',text,flags=re.I)
    text = re.sub(r'\ba+n+\b','an',text,flags=re.I)
    text = re.sub(r'\ba+n+d+\b','and',text,flags=re.I)
    text=re.sub(r'\bab+andon','abandon',text,flags=re.I)
    return text
from emo_dictonary import EMOTICONS
def remove_emoticon(text):
    for emot in EMOTICONS:
        if emot[0] in string.punctuation:
            if emot[-1] in string.punctuation:
                emoticon_pattern = r'\B'+emot+r'\B'
            else:
                emoticon_pattern = r'\B'+emot+r'\b'
        else:
            if emot[-1] not in string.punctuation:
                emoticon_pattern = r'\b'+emot+r'\b'
            else:
                emoticon_pattern = r'\b'+emot+r'\B' 
        text = re.sub(emoticon_pattern," ", text,flags=re.I)
    return text

slang_expressions = pd.read_csv('slang_words.csv', sep=',')
slang_expressions_dict = dict(zip(slang_expressions.Slang_key, slang_expressions.Slang_value))
def slang_cleaner(text):
    for key in slang_expressions_dict:
        text=re.sub((r""+key),(r""+slang_expressions_dict[key]),text)
    return text
def slang_spelling(text):
    text = re.sub(r"\bUK\b", "United Kingdom", text)
    text = re.sub(r"recentlu", "recently", text)
    text = re.sub(r"Ph0tos", "Photos", text)
    text = re.sub(r"amirite", "am I right", text)
    text = re.sub(r"exp0sed", "exposed", text)
    text = re.sub(r"amageddon", "armageddon", text)
    text = re.sub(r"TRAUMATISED", "traumatized", text)
    text = re.sub(r"2k([0-9])([0-9])", r"20\1\2", text,flags=re.I)
    text = re.sub(r"Trfc", "Traffic", text)
    text=re.sub(r"\bbout\b","about",text)
    text=re.sub(r"\bcsl\b","can not stop laughing",text)
    text=re.sub(r"\bcyt\b","see you tomorrow",text)
    text=re.sub(r"\bidk\b","i do not know",text)
    text=re.sub(r"\bsmh\b","shake my head",text,flags=re.IGNORECASE)
    text = re.sub(r"\bgon\b", r"going to", text,flags=re.I)
    text = re.sub(r"\bwut\b", r"what", text,flags=re.I)
    text = re.sub(r"\byday\b", r"yesterday", text,flags=re.I)
    text = re.sub(r'\baccidently\b',"accidentally",text)
    text = re.sub(r"\bStree\b",'street',text)
    text = re.sub(r"\blil\b",' little ',text)
    text = re.sub(r"\balil\b",' a little ',text)
    text = re.sub(r"\btxt\b",' text',text,flags=re.I)
    text = re.sub(r"\btyre\b",'tire',text,flags=re.I)
    text = re.sub(r"\bknw\b",'know',text,flags=re.I)
    text = re.sub(r"\bER\b","emergency room",text)
    text = re.sub(r"\bS3XLEAK",'SEX LEAK',text)
    text = re.sub(r"view and download video",' ',text)
    text = re.sub(r"\bvia\b",' ',text,flags=re.I)
    text=re.sub(r"\bygtr\b","you got that right",text,flags=re.I)
    text=re.sub(r"\bwuzup\b","what is up",text,flags=re.I)
    text=re.sub(r"\byd\b","yard",text,flags=re.I)
    text=re.sub(r"\bzzz\b","sleeping bored and tired",text)
    text=re.sub(r"\btho\b","though",text)
    text=re.sub(r"\btia\b","thanks in advance",text)
    text=re.sub("\butc","coordinated universal time ",text)
    text=re.sub("\bsrsly\b","seriously",text)
    text=re.sub(r"\bRT((\?)?)",r"retweet \1",text)
    text=re.sub("\bqpsa\b","what happens",text)
    text=re.sub("\bps\b","postscript",text)
    text=re.sub("\bpov\b","point of view",text,flags=re.I)
    text=re.sub("\bnhs\b","national health service",text,flags=re.I)
    text=re.sub("\bnrn\b","no reply necessary",text,flags=re.I)
    text=re.sub(r"\bAMA\b","ask me anything",text)
    text=re.sub(r"\bbe4\b","before",text)
    text=re.sub(r"\bbbc\b","british broadcasting corporation",text,flags=re.I)
    text=re.sub(r"\bdm\b","direct message",text,flags=re.I)
    text=re.sub(r"\betc\b","and so on",text,flags=re.I)
    text=re.sub(r"\bft.\B","feet",text)
    text=re.sub(r"\bgmt\b","greenwich mean time",text,flags=re.I)
    text=re.sub(r"\bgps\b","global positioning system",text,flags=re.I)
    text=re.sub(r"\bidc\b","i do not care",text,flags=re.I)
    text=re.sub(r"\bidgaf\b","i do not give a fuck",text,flags=re.I)
    text=re.sub(r"\bimo\b","in my opinion",text,flags=re.I)
    text=re.sub(r"\blb(s?)\b",r"pound\1",text,flags=re.I)
    text=re.sub(r"\bnvr\b","never",text,flags=re.I)
    text=re.sub(r"\btbh\b","to be honest",text,flags=re.I)
    text=re.sub(r"\bpov\b","point of view",text,flags=re.I)
    text=re.sub(r"\B\[IR\]\B",r"\[Information reveal\]",text)
    text=re.sub(r"\besp\.\B",r"especially",text)
    text=re.sub(r"\bcuz\b",r"because",text)
    text=re.sub(r"\bcya\b",r"see you",text)
    text=re.sub(r"\bplz\b",r"please",text)
    text=re.sub(r"\b(\d+)(\s?)yr(s?)\b",r"\1 year\3",text)
    #text=re.sub(r"\b(\d+)(\s?)[Kk]m\b",r"\1 kilometers",text)
    text=re.sub(r"\bb4federal\b",r"before federal",text)
    #text=re.sub(r"\blmfao\b",r"laugh my fucking ass off",text)
    return text
def clean_contractions(text):
    text=contractions.fix(text)
    text = re.sub(r"\bcant\b",'can not',text)
    text = re.sub(r"\bwont\b",'will not',text)
    text = re.sub(r"\bim\b",' I am ',text)
    text = re.sub(r"\bdidnt\b",' did not ',text)
    text = re.sub(r"\bcouldnt\b",'could not',text)
    text = re.sub(r"\bisnt\b",' is not ',text)
    text = re.sub(r"\bdont\b",' do not ',text)
    return text
# def remove_irr_char_func(text):
#     text=re.sub(r'[^a-zA-Z]', ' ', text)
#     return text
def remove_punctuation_func(text):
    return re.sub(r'[^a-zA-Z0-9]', ' ', text)
chat_expressions = pd.read_csv('chat_expressions.csv', sep=',')
chat_expressions_dict = dict(zip(chat_expressions.Chat_Words, chat_expressions.Chat_Words_Extended))
def chat_words_to_norm_words_func(text):
    return re.sub(r'\S+', lambda m: chat_expressions_dict.get(m.group().upper(), m.group()) , text)
def remove_extra_whitespaces_func(text):
    return re.sub(r'^\s*|\s\s*', ' ', text).strip()
def clean_text(text):
    text=remove_html(text)
    text=remove_url_func(text)
    text=remove_tweet_funct(text)
    text=remove_accented_chars_func(text)
    text=clean_repeating_char(text)
    text=remove_emoticon(text)
    text=slang_cleaner(text)
    text=slang_spelling(text)
    text=clean_contractions(text)
    text=remove_punctuation_func(text)
    text=chat_words_to_norm_words_func(text)
    text=remove_extra_whitespaces_func(text)
    return text
stop_words = set(stopwords.words('english'))
wn = nltk.WordNetLemmatizer()
worddict = set(nltk.corpus.words.words())
def preprocessing(text):
    new_text = clean_text(text)
    tokens = word_tokenize(new_text.lower().strip())
#     new_text=" ".join([wn.lemmatize(w) for w in tokens if not w in stop_words])
#     return new_text
    wordset_n = set(wn.lemmatize(w, NOUN) for w in tokens if not w in stop_words)
    wordset_v = set(wn.lemmatize(w, VERB) for w in wordset_n)
    wordset = set(wn.lemmatize(w, ADJ) for w in wordset_v)
    wordset = wordset & worddict
    return ' '.join(list(wordset))
```

```python
data_train['clean_text']=data_train['text'].apply(lambda x: clean_text(x))
data_test['clean_text']=data_test['text'].apply(lambda x: clean_text(x))
data_train['new_text']=data_train['text'].apply(lambda x: preprocessing(x))
data_test['new_text']=data_test['text'].apply(lambda x: preprocessing(x))
data_train.to_csv('tem_train.csv',encoding='ISO-8859–1',index=False)
data_test.to_csv('tem_test.csv',encoding='ISO-8859–1',index=False)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    data_train['new_text'],
    data_train['target'],stratify=data_train['target'],
    test_size=0.25,
    random_state=2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics
tf_transformer=TfidfVectorizer(use_idf=True)
x_train_tf = tf_transformer.fit_transform(x_train)
y_train = np.array(y_train)
x_pre_tf = tf_transformer.transform(x_test)
x_test_tf = tf_transformer.transform(data_test['new_text'])
```

```python
for item in tf_transformer.get_feature_names_out():
    print(item)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
MNB_model = MultinomialNB(alpha=1.06, fit_prior=True, class_prior=None)
MNB_model.fit(x_train_tf, y_train)
predicted = MNB_model.predict(x_pre_tf)
c5=metrics.classification_report(list(y_test), predicted)
print(c5)
```

```python
cm = confusion_matrix(list(y_test), predicted)
cm
```

```python
metrics.accuracy_score(list(y_test), predicted)
```

```python
metrics.f1_score(list(y_test), predicted)
```
